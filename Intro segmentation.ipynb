# Introduction à la segmentation d'image

# 1 - Définitions

- La segmentation d'image est une méthode qui consiste à délimiter de façon automatique de zones de pixels dans une image. Ces zones de pixels sont sensés appartenir à une même classe d'objets. 


- La segmentation d'image s'utilise courramment pour la détection automatique d'objets. Au lieu de traiter l'image entière, on utilise d'abord un algorithme de segmentation pour trouver des objets d'intèrêt dans l'image. Cela permet à l'algorithme de detection d'objet de ne pas traiter entièrement l'image, ce qui améliore la précision et réduit le temps de calcul. 


- La segmentation d'image est utile pour l'analyse d'image médicales, la création de véhicules autonomes, la reconnaissance et la détection de visage, la vidéosurveillance et l'analyse d'images satéllitaires. 

# 2 - Les méthodes de segmentation d'image

## 2 - 1 - La segmentation de contours

Elle identifies les bords (ou contours) de divers objet dans une image

## 2 - 2 La segementation oar seuillage


Elle catégorise les pixels d'une image en 2 partitions séparées par une valeur seuil de pixel. Elle s'utilise lorsque l'objectif de la segmentation est de produire une séparation entre l'objet d'intèrêt et l'arrière plan de l'image 

## 2 - 3 La segmentation en régions 


Elle partitionne l'image en plusieurs régions présentant des caratéristiques similaires
